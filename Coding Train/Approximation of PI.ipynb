```python
import matplotlib.pyplot as plt
import random
import math
import pylab
circle_radius  = 1.0
sqLowerX = -1.0
sqLowerY = -1.0
randomX = 0.0
randomY = 0.0
```

```python
def setup():

    print ("setup")
    plt.axes()
    # Add a rectangle
    rectangle = plt.Rectangle((sqLowerX, sqLowerY), circle_radius*2, circle_radius*2, 
                              fc='w', edgecolor='b')
    plt.gca().add_patch(rectangle)

    
    # Add a Circle
    circle = plt.Circle((0, 0), color='r', alpha=0.5, radius=circle_radius, fc='w')
    plt.gca().add_patch(circle)

    plt.axis('scaled')
    
    axes = plt.gca()
    axes.set_xlim([-1.2,1.2])
    axes.set_ylim([-1.2,1.2])
    
```

```python
def draw(limit):
    i = 0
    totalCount = 0
    circleCount = 0
    my_pi_list = list()
    
    while (i < limit):
        
        i += 1
        
        for x in range(10):
            randX = float(random.uniform(-circle_radius,circle_radius))
            randY = float(random.uniform(-circle_radius,circle_radius))

            totalCount += 1

            d_SQR = float(math.pow(randX,2)) + float(math.pow(randY,2))

            if d_SQR < math.pow(circle_radius, 2):
                plt.plot(randX,randY,'go', markersize=1)
                circleCount += 1
            else :
                plt.plot(randX,randY,'bo',markersize=1)

            pi = 4 * (circleCount / totalCount)
            print(pi, end='\r')
            #my_pi_list.append(pi)
        
    plt.show()
    #print(my_pi_list,end='\r')
```

```python
multipler = 10
graphs = 1
while graphs <= 5:
    multipler = multipler * 10
    setup()
    draw(multipler)
    print("-------------------", end='\n')
```
